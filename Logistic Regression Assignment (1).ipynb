# 2. Data Preprocessing:

train _data

filling cabin with unknown

filling embarked with mode

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
```

```python
train_data['Cabin'].fillna('Unknown', inplace=True)
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Cabin'].fillna('Unknown', inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

# 3. Model Building:
a. Build a logistic regression model using appropriate libraries (e.g., scikit-learn).

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression().fit(x_train,y_train)
logistic
```

b. Train the model using the training data.

# predicting the on train data

```python
train_pred=logistic.predict(x_train)
train_pred
```

# predicting the on test data

```python
test_pred=logistic.predict(test)
test_pred
```

```python
y_pred_df=pd.DataFrame({
                        'predicted_prob':logistic.predict_proba(test)[:,1],
                        'y_predictons':test_pred})
y_pred_df
```

# 4. Model Evaluation:
#  Interpret the coefficients of the logistic regression model.

```python
#coefficients for each feature
coefficients = logistic.coef_[0]  # Coefficients for each feature
intercept = logistic.intercept_[0] # Intercept (bias term)
print('coefficients:',coefficients)
print('intercept:',intercept)
```

```python
for feature, coef in zip(x_train.columns, coefficients):
    print(f"{feature}: {coef:.4f}")
print(f"Intercept: {intercept:.4f}")
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np
import pandas as pd  # Add the missing import

# Load the trained model
with open("survival_model.pkl", "rb") as file:
    model = pickle.load(file)

# App title
st.title("Titanic Survival Prediction App")
# Sidebar inputs
st.sidebar.header("Passenger Details")
Pclass = st.sidebar.selectbox("Ticket Class (1 = First, 2 = Second, 3 = Third):", [1, 2, 3])
Age = st.sidebar.slider("Age:", 0, 100, 30)
Fare = st.sidebar.slider("Fare:", 0.0, 500.0, 50.0)
SibSp = st.sidebar.number_input("Number of Siblings/Spouses Aboard:", 0, 10, 0)
Parch = st.sidebar.number_input("Number of Parents/Children Aboard:", 0, 10, 0)
Sex_male = st.sidebar.selectbox("Sex:", ["Female", "Male"])
Sex_male = 1 if Sex_male == "Male" else 0
Embarked_S = st.sidebar.selectbox("Port of Embarkation:", ["C", "Q", "S"])
Embarked_S = 1 if Embarked_S == "S" else 0
Embarked_C = 1 if Embarked_S == "C" else 0
Embarked_Q = 1 if Embarked_S == "Q" else 0
Pclass_1 = 1 if Pclass == 1 else 0
Pclass_2 = 1 if Pclass == 2 else 0
Pclass_3 = 1 if Pclass == 3 else 0
Sex_female = 1 if Sex_male == 0 else 0  # Female is the opposite of male

# Create input dataframe
input_data = pd.DataFrame([[Age, SibSp, Parch, Fare, Pclass_1, Pclass_2, Pclass_3,
                            Embarked_C, Embarked_Q, Embarked_S, Sex_female, Sex_male]],
                          columns=['Age', 'SibSp', 'Parch', 'Fare', 'Pclass_1', 'Pclass_2', 'Pclass_3',
                                   'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Sex_female', 'Sex_male'])

# Prediction button
if st.button("Predict"):
    prediction = model.predict(input_data)
    if prediction[0] == 1:
        st.success("The passenger would have survived!")
    else:
        st.error("The passenger would not have survived.")
```

Why Is This Useful in Binary Classification?

In binary classification (e.g., predicting “Yes” or “No”):

Fair Testing: It ensures every data point gets tested once.

Better Performance Measures:Metrics like accuracy, precision, and recall might vary based on how the data is split.
